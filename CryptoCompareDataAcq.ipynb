```python
#function to get historical prices for each symbol
def getpricedflist(sym):
    dic = ccom.get_historical_price_day(sym, 'USD', limit=1200, exchange='CCCAGG', toTs=datetime.date.today())
    df = pd.DataFrame.from_dict(dic, orient='columns')
    df['Symbol'] = sym
    dflist.append(df)
```

```python
#Filter symbols
coinsin = list(set(df7.Symbol))
coinsall = coinsin + coinsnan
coins_filt = coins_ser[~coins_ser.isin(coinsall)]
```
